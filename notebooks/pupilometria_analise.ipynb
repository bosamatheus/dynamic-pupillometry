Importando as bibliotecas utilizadas neste experimento.

```
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("ticks")

import numpy as np
import pandas as pd
```

Carregando arquivo csv contendo os resultados para um paciente.

*   Paciente: 12
*   Coleta: 2

```
path = '/content/drive/My Drive/TCC/Scripts/Resultados/'
patient = '12_1'
#script = '_hist'
script = '_grow'

filename = path + 'results_' + patient + script + '.csv'

df = pd.read_csv(filename, sep=',');
df
```

Visualizando a resposta da pupila e da íris ao longo do tempo.

```
flash_step_time = 3*10e-3
N = 49

t = np.linspace(0, flash_step_time*N, N)

plt.figure(figsize=(14, 8))
plt.title('Resposta temporal da pupila e da íris')
plt.xlabel('Tempo (segundos)')
plt.ylabel('Pixels')
plt.plot(t, df['radius_pupil'], '--ro', label='Raio da pupila')
plt.plot(t, df['radius_iris'], '--bo', label='Raio da íris')
plt.legend()
plt.grid()
plt.show()
```
